```python
## Importing the required libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
## Reading the data

df = pd.read_csv('ecommerce_sellers.csv')
df_quality = pd.read_csv('noise\ecommerce_sellers_with_best_quality_with_noise.csv')
df_time = pd.read_csv('noise\ecommerce_sellers_with_quick_delivery_with_noise.csv')
df_cost = pd.read_csv('noise\ecommerce_sellers_with_lowest_cost_with_noise.csv')
```

```python
df_features = df.drop(['category'], axis=1)
features = np.array(df_features)
df_features.head(5)
```

```python
target_cost = df_cost['lowest_cost']
target_time = df_time['quick_delivery']
target_quality = df_quality['best_quality']

cost = np.array(target_cost)
time = np.array(target_time)
quality = np.array(target_quality)
```

```python
import seaborn as sns

## plotting every column in the features matrix against the target column

for i in range(0, 8):
    sns.scatterplot(x=df_features.columns[i], y='lowest_cost', data=df_cost)
    plt.show()
```

```python
## Splitting the data into training and testing sets

X_train_cost, X_test_cost, y_train_cost, y_test_cost = train_test_split(features, cost, test_size=0.2, random_state=42)
X_train_time, X_test_time, y_train_time, y_test_time = train_test_split(features, time, test_size=0.2, random_state=42)
X_train_quality, X_test_quality, y_train_quality, y_test_quality = train_test_split(features, quality, test_size=0.2, random_state=42)
```

```python
## importing regression models

from sklearn.linear_model import LinearRegression, ElasticNet, SGDRegressor, Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor

from sklearn.metrics import mean_squared_error, r2_score
```

```python
## Linear Regression

lr_cost = LinearRegression()
lr_cost.fit(X_train_cost, y_train_cost)
lr_cost_pred = lr_cost.predict(X_test_cost)
lr_cost_mse = mean_squared_error(y_test_cost, lr_cost_pred)
lr_cost_r2 = r2_score(y_test_cost, lr_cost_pred)

print('Linear Regression Cost MSE: ', lr_cost_mse)
print('Linear Regression Cost R2: ', lr_cost_r2)
```

```python
## Ridge Regression

ridge_cost = Ridge(alpha=0.001, fit_intercept=True, copy_X=True, max_iter=None, solver='cholesky', positive=False, random_state=None)
ridge_cost.fit(X_train_cost, y_train_cost)
ridge_cost_pred = ridge_cost.predict(X_test_cost)
ridge_cost_mse = mean_squared_error(y_test_cost, ridge_cost_pred)
ridge_cost_r2 = r2_score(y_test_cost, ridge_cost_pred)

print('Ridge Regression Cost MSE: ', ridge_cost_mse)
print('Ridge Regression Cost R2: ', ridge_cost_r2)
```

```python
## Elastic Net

en_cost = ElasticNet()
en_cost.fit(X_train_cost, y_train_cost)
en_cost_pred = en_cost.predict(X_test_cost)
en_cost_mse = mean_squared_error(y_test_cost, en_cost_pred)
en_cost_r2 = r2_score(y_test_cost, en_cost_pred)

print('Elastic Net Cost MSE: ', en_cost_mse)
print('Elastic Net Cost R2: ', en_cost_r2)
```

```python
## SGD Regressor

sgd_cost = SGDRegressor()
sgd_cost.fit(X_train_cost, y_train_cost)
sgd_cost_pred = sgd_cost.predict(X_test_cost)
sgd_cost_mse = mean_squared_error(y_test_cost, sgd_cost_pred)
sgd_cost_r2 = r2_score(y_test_cost, sgd_cost_pred)

print('SGD Regressor Cost MSE: ', sgd_cost_mse)
print('SGD Regressor Cost R2: ', sgd_cost_r2)
```

```python
## Random Forest Regressor

rf_cost = RandomForestRegressor(n_estimators=1000, max_features='sqrt')
rf_cost.fit(X_train_cost, y_train_cost)
rf_cost_pred = rf_cost.predict(X_test_cost)
rf_cost_mse = mean_squared_error(y_test_cost, rf_cost_pred)
rf_cost_r2 = r2_score(y_test_cost, rf_cost_pred)

print('Random Forest Regressor Cost MSE: ', rf_cost_mse)
print('Random Forest Regressor Cost R2: ', rf_cost_r2)
```

```python
## SVR

svr_cost = SVR(C=1e5, epsilon=2e-5)
svr_cost.fit(X_train_cost, y_train_cost)
svr_cost_pred = svr_cost.predict(X_test_cost)
svr_cost_mse = mean_squared_error(y_test_cost, svr_cost_pred)
svr_cost_r2 = r2_score(y_test_cost, svr_cost_pred)

print('SVR Cost MSE: ', svr_cost_mse)
print('SVR Cost R2: ', svr_cost_r2)
```

```python
## Decision Tree Regressor

dt_cost = DecisionTreeRegressor()
dt_cost.fit(X_train_cost, y_train_cost)
dt_cost_pred = dt_cost.predict(X_test_cost)
dt_cost_mse = mean_squared_error(y_test_cost, dt_cost_pred)
dt_cost_r2 = r2_score(y_test_cost, dt_cost_pred)

print('Decision Tree Regressor Cost MSE: ', dt_cost_mse)
print('Decision Tree Regressor Cost R2: ', dt_cost_r2)
```

__Finally, ofc linear regression works best so that's the final model!__

```python
## re-training on the entire dataset

lr_cost = LinearRegression()
lr_cost.fit(features, cost)

lr_time = LinearRegression()
lr_time.fit(features, time)

lr_quality = LinearRegression()
lr_quality.fit(features, quality)
```

```python
## pickle the final models

import pickle

pickle.dump(lr_cost, open('lr_cost.pkl', 'wb'))
pickle.dump(lr_time, open('lr_time.pkl', 'wb'))
pickle.dump(lr_quality, open('lr_quality.pkl', 'wb'))
```
